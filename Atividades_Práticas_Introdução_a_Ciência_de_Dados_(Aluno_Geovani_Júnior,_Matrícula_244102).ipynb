```python
# Definindo variáveis

fatias = [8, 12, 10]

atividades = ['Lazer', 'Estudo', 'Viagem']

colunas = ['r', 'm', 'y']

# Criando um Gráfico

plt.pie(fatias, labels = atividades, colors = colunas, startangle = 90,
shadow = True, explode = (0, 0.1, 0))

plt.show()
```

```python
# Definindo variáveis

x = [1, 3, 5]

y = [1, 2, 5]

# Criando um Gráfico

plt.plot(x, y)

# Atribuindo um título ao gráfico

plt.title('Exemplo utilizando Plot')

plt.xlabel('Variavel 1')

plt.xlabel('Variavel 2')

# Atribuindo uma legenda

plt.plot(x, y, label = 'Uma legenda')

plt.legend()

# Exibindo o gráfico gerado

plt.show()
```

```python
# Definindo variáveis

x = [1, 2, 3, 4, 5, 6, 7, 8]

y = [5, 2, 4, 5, 6, 8, 4, 8]

# Criando um gráfico

plt.scatter(x, y, label = 'Pontos', color = 'b', marker = '*', s = 100)

plt.legend()

plt.show()
```
